```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("pikachuu.jpeg")
```

```python
img2= cv2.imread("candy.jpeg")
img2= img2[50:50+230, 50:50+219,:]
```

```python
img.shape, img2.shape
```

```python
def imshow(im):
    cv2.imshow("preview", im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
imshow(img)
imshow(img2)
```

```python
def showing(im):
    plt.imshow(im[:,:,::-1])
    plt.show()
```

```python
showing(img)
```

```python
showing(img2)
```

```python
added=cv2.add(img,img2)
showing(added)

# feature similar to increasing brightness in the phone
```

```python
subtract=cv2.subtract(img,img2)
showing(subtract)


# brightness kam krne k liye
```

```python
img_=cv2.subtract(img, 20*np.ones(img.shape).astype('uint8'))
showing(img_)
```

```python
#0 - 255 uint k constarint k sth 0 dega

# contrast bhadane k liye we multiply the image
mul=cv2.multiply(img,img2)
showing(mul) 
```

```python
mul=cv2.multiply(img, 2*np.ones(img.shape).astype('uint8'))
showing(mul)
```

```python
# contrast kam krne k liye

div=cv2.divide(img,img2)
showing(div)
```

```python
img_2=cv2.convertScaleAbs(img,alpha=1.1,beta=80)   # brightness, contrast change
showing(img_2)
```

```python
# re sizing image

img_3=cv2.resize(img2,(400,400),interpolation=cv2.INTER_CUBIC)
img_3.shape
```

```python
showing(img_3)
```

```python
img_4=cv2.resize(img2, None, fx=2, fy=0.5, interpolation=cv2.INTER_CUBIC)
img_4.shape
```

```python
showing(img_4)
```

```python
from skimage.transform import rotate
```

```python
# rotation

img_rotate = rotate(img, angle=40, resize=True)
imshow(img_rotate)
```

```python
showing(img_rotate)
```

```python
from time import sleep
repeat= True
while repeat:
    for i in range(0,361,1):  # or 360
        i=rotate(img, angle=i, resize=True)
        cv2.imshow("", i)
        key= cv2.waitKey(1)
        if key== ord('x'):
            repeat=False
            break
        sleep(0.1)
cv2.destroyAllWindows()
```
