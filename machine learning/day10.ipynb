```python
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
ski.io.imshow(img_gray)
```

```python
img_gray1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ski.io.imshow(img_gray1)
```

```python
#img into binary
_, img_binary=cv2.threshold(img_gray,50,255,cv2.THRESH_BINARY)
ski.io.imshow(img_binary,)
```

```python
_, img_binary=cv2.threshold(img_gray,50,255,cv2.THRESH_BINARY_INV)
ski.io.imshow(img_binary,)
```
